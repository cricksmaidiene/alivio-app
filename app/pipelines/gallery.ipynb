```python
import boto3
import pandas as pd
```

```python
s3 = boto3.client('s3')
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_parquet("s3://alivio/datasets/xview2/processed/images.parquet")

df = df[~df.is_pre_image]
df = df[df.disaster.str.contains("hurricane", case=False)]

df["image_filepath"] = df.apply(
    lambda row: f"s3://alivio/datasets/xview2/challenge/{row['dataset']}/images/{row['img_name']}",
    axis=1,
)
df["json_filepath"] = df.apply(
    lambda row: f"s3://alivio/datasets/xview2/challenge/{row['dataset']}/labels/{row['img_name'].replace('.png', '.json')}",
    axis=1,
)

print(df.shape)
df.head()
```

```python
df.to_parquet("s3://alivio/datasets/xview2/processed/app_gallery_dataset.parquet", index=False)
```
